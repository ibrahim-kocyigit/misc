# Stage 3: Data Requirements

In this stage, we define what data is needed to train our baseline and candidate models. This includes identifying the data sources and required data fields, specifying the format, granularity, and timeframe of the data, and identifying the data privacy and ethical considerations.

---

## 1. Target Variable Definition
* **Name:** `is_versicolor`
* **Type:** Binary (0/1)
* **Definition:** 1 if the iris flower is of the species Versicolor, 0 otherwise.
* **Known Issues:** The dataset does not contain `is_versicolor` as a direct field; it must be derived from the `species` field.

## 4. Format & Granularity
* **Granularity:** One row per iris sample.
* **Timeframe:** All iris samples are from a single collection period; no temporal component is present in the dataset.
* **Format:** One CSV file, comma-delimited, UTF-8 encoded.
* **Expected Volume:** ~50,000 observations

---

## 5. Privacy & Ethics Review
* **PII Fields Identified:** The dataset contains only measurements of iris flowers and their species, which are not personally identifiable information (PII). Therefore, there are no PII fields to list or handle in this project.
* **Legal Constraints:** None.
* **Authorization:** Client authorized use of this data for this project via email on 2026-03-17.
