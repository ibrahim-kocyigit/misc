# Stage 2: Analytic Approach

In this stage we translate the business problem from [Stage 1: Business Understanding](1.01-business_understanding.ipynb) into an analytical framework. We answer: "What type of data science problem is this, what algorithms could solve it, and how will we measure whether they work?"

> **Dataset Note:** This project uses a [dataset](https://www.kaggle.com/datasets/drrayislam/50k-synthetic-iris-data-set-49736-observations) of approximately 50,000 labeled Iris observations. The dataset was created by Dr. Ray Islam by generating around 50K synthetic data sets using CTGAN resembling the real data set. For the purposes of this tutorial, we treat this as a real historical record consistent with the business narrative established in Stage 1.

---

## 1. ML Problem Framing
* **Type of Problem:** Supervised Binary Classification
* **Target Variable:** `is_versicolor`
* **Justification:**
    * The dataset contains labeled examples, which makes this a **supervised** learning setup where the model learns from known outcomes.
    * The stakeholders need a yes or no decision for each flower — *versicolor* or *not versicolor* — so the task is inherently **binary**.
    * The predictors (sepal and petal measurements) are suitable inputs for **classification** algorithms. This framing also supports the business priority that false versicolor predictions are more costly than missed versicolor cases, because model thresholding and evaluation can be tuned to minimize that specific error type.

---

## 2. Candidate Models

1. **Baseline: Logistic Regression.** Fast and computationally cheap. A reliable first benchmark before trying more complex models. Even though interpretability is not a priority for the stakeholders, the interpretable coefficients the model provides can help us understand the relationship between features and the target variable, which is useful for debugging and feature engineering.
2. **Candidate Model 1: Random Forest.** A good candidate for general-purpose classification when we want high accuracy with minimal preprocessing and hyperparameter sensitivity. Typically less prone to overfitting than single decision trees. Also provides feature importance scores, which can help us understand which features are most influential in predicting versicolor.
4. **Candidate Model 2: Support Vector Machine (SVM).** Can model both linear and non-linear decision boundaries using kernel functions. Class weights can be adjusted to align with the business priority of minimizing false versicolor predictions. Requires feature scaling and careful kernel and hyperparameter selection.
3. **Candidate Model 3: XGBoost.** An efficient implementation of gradient boosting that often achieves strong performance on structured tabular data. Captures complex relationships and interactions between features, and includes built-in regularization to reduce overfitting.

> **Preprocessing Note:** Logistic Regression and SVM require feature scaling to perform correctly. Tree-based models (Random Forest, XGBoost) do not. Each candidate model will be evaluated inside its own preprocessing pipeline to ensure a fair and unbiased comparison.

* **Interpretability Note:** The business owner does not prioritize interpretability, so we are not constrained to interpretable models. We will still consider interpretability as a secondary factor — particularly where feature importance scores can support debugging or help validate the model's behavior against domain knowledge.

---

## 3. Evaluation Metrics

* **Primary Metric: Precision (Versicolor class).** The business owner has emphasized that false versicolor predictions are far more costly than missed versicolor cases — a false positive sends the wrong flower into the drug pipeline. Precision directly measures the share of predicted versicolor flowers that are truly versicolor, which is the clearest expression of this business risk.

* **Secondary Metric: Precision-Recall AUC (PR AUC).** Provides a comprehensive view of model performance across all classification thresholds. This is the key metric for comparing candidate models holistically and for selecting the optimal operating threshold.

* **Acceptance Threshold:** We target a versicolor **precision of at least 0.90** alongside a minimum versicolor **recall of 0.70**. The precision floor reflects the business cost of false versicolor predictions. The recall floor ensures the system identifies a sufficient proportion of true versicolor flowers and does not over-filter to artificially inflate precision. These thresholds are not rigid and will be reviewed with the business owner once candidate model results are available.

---

## 4. Validation Strategy

* **Split:** 80/20 stratified split to preserve class proportions across training and test sets.
* **Cross-Validation:** 5-fold cross-validation on the training set for robust model selection and threshold tuning across multiple data splits.
* **Final Evaluation:** The selected model will be retrained on the full training set and evaluated once on the holdout test set. These are the final reported metrics.
