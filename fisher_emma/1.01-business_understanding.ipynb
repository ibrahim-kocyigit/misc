## 2. Business Problem Statement

The Fisher Botanical Research Institute has been contracted by NovaCure Pharmaceuticals to identify and sort thousands of *Iris versicolor* specimens for pharmaceutical production within a 3-month deadline. The current manual identification process — performed by only two expert botanists at approximately 5 minutes per flower — cannot scale to meet the required volume of 2,000–3,000 flowers per week. The Institute needs an automated classification system that enables quickly-trained interns to identify *Iris versicolor* from simple physical measurements, eliminating the expertise bottleneck and meeting NovaCure's production timeline.

---

## 3. Project Objectives & Success Criteria

1. **Primary Objective:** Build a classification system that predicts whether a given Iris flower is *versicolor* or *not versicolor* based on four physical measurements (sepal length, sepal width, petal length, petal width), enabling non-expert interns to perform species identification in seconds rather than minutes.
2. **Secondary Objective(s):**
    * Provide a simple, user-friendly web interface where interns can input measurements and receive an instant prediction.
    * Expose an API endpoint so that NovaCure's internal systems can programmatically request classifications if they decide to scale or automate their sorting line.
    * Implement a lightweight logging mechanism that records predictions so the quality team can periodically export and review them against expert spot-checks.
3. **Success Criteria:**
    * The system must be operational and deployed before the end of the 3-month window (target: **by 2026-06-13**).
    * The classification must **minimize false versicolor predictions** — flowers incorrectly classified as *versicolor* must be rare, even if it means some true *versicolor* specimens are missed. (The exact metric threshold will be defined in Stage 2: Analytic Approach.
    * Interns with no botanical expertise should be able to use the tool after a brief orientation (under 30 minutes).
    * The total identification time per flower (measure + input + prediction) should be under **2 minutes**, representing at least a 60% improvement over the current 5-minute expert process.
    * Periodic prediction logs must be available for export so the quality team can compare predictions against expert verification samples on a weekly/bi-weekly basis.

---

## 4. Solution Requirements

* **Output:** A binary prediction — ***versicolor*** or ***not versicolor*** — for each set of four input measurements (sepal length, sepal width, petal length, petal width). The result should be presented clearly and unambiguously to the intern.
* **Delivery:**
    * **Primary:** A web application where interns manually enter the four measurements and receive an instant prediction.
    * **Secondary:** An API endpoint that accepts measurement inputs and returns the predicted class, enabling NovaCure to integrate classification into their internal systems if they choose to automate.
* **Interpretability:** Not critical. The stakeholders have explicitly stated that a black-box prediction is acceptable given the time constraints. Understanding *why* a specific classification was made is not required.
* **Cadence:** Ongoing — the system will be used continuously throughout the 3-month sorting campaign. This is a deployed, always-available prediction service, not a one-time analysis.
* **Logging:** The system must log each prediction (input measurements, predicted class, timestamp). Logs will be exported periodically (weekly or bi-weekly) as simple report files for the quality team to review against expert spot-check results. This is for accuracy monitoring, not legal compliance.
